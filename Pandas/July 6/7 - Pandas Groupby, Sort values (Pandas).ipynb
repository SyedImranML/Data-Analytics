### Combining & Merging Datasets in Pandas

```python
import pandas as pd
import numpy as np
```

### Joining DataFrame

```python
d1=pd.DataFrame(
       {"key":["a","b","c","c","d","e"],
        "num1":range(6)})
d2=pd.DataFrame(
       {"key":["b","c","e","f"],
        "num2":range(4)})
```

```python
print(d1)
print('-'*20)
print(d2)
```

```python
pd.merge(d1,d2,on='key')
```

Keys are Different

```python
d3=pd.DataFrame(
       {"key1":["a","b","c","c","d","e"],
        "num1":range(6)})
d4=pd.DataFrame(
       {"key2":["b","c","e","f"],
        "num2":range(4)})
```

```python
d3.head()
```

```python
d4.head()
```

```python
pd.merge(d3,d4,left_on='key1',right_on='key2',how = 'inner')
```

```python
pd.merge(d3,d4,left_on='key1',right_on='key2',how = 'left')
```

```python
d3=pd.DataFrame(
       {"key1":["a","b","c","c","d","e"],
        "num1":range(6)})
d4=pd.DataFrame(
       {"key2":["b","c","e","f"],
        "num1":range(4)})
```

```python
pd.merge(d3,d4,left_on='key1',right_on='key2',how = 'left') #observation the column gets changed as num_x,num_y
```

```python
pd.merge(d1,d2,on='key',how='outer')
```

```python
pd.merge(d1,d2,on='key',how='left')
```

```python
pd.merge(d1,d2,on='key',how='right')
```

```python
pd.merge(d1,d2,on='key',how='inner')
```

```python
df1=pd.DataFrame(
   {"key":["a","b","c","c","d","e"],
   "num1":range(6),
   "count":["one","three","two","one","one","two"]})

df2=pd.DataFrame(
   {"key":["b","c","e","f"],
   "num1":range(4),
   "count":["one","two","two","two"]})
```

```python
df1.head()
```

```python
df2.head()
```

```python
pd.merge(df1,df2,left_on=["key","count"],
        right_on =["key","count"])
```

```python
pd.merge(df1,df2,on=["key","count"])
```

```python
d3=pd.DataFrame(
       {"key1":["a","b","c","c","d","e"],
        "num1":range(6)})
d4=pd.DataFrame(
       {"key2":["b","c","e","f"],
        "num1":range(4)})
```

```python
pd.merge(d3,d4,left_on="key1",right_on="key2",how='inner')
```

```python
pd.merge(d3,d4,left_on="key1",right_on="key2",how='inner',
        suffixes=('_data1','_data2'))
```
